```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

plot is the basic plot i.e. line plot in matplotlib, 
Font size and style can be adjusted for Title names. 
Graph will adjust as per ticks.
Label works only after legend is mentioned in the code.


```python
x=[1,2,3]
y=[2,4,6]

plt.plot(x,y, label='x to y', color='yellow', linewidth=4, marker='.', markeredgecolor='orange', linestyle='--')
plt.title('First Graph', fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('Y axis')

plt.xticks([1,2,3])
plt.yticks([2,4,6])

plt.legend()


plt.show()
```

```python
x2=np.arange(1,7,0.7)

#dpi = pixel by inch

plt.figure(figsize=(5,3), dpi=100)

#short hand notation for color, marker and line is 'b.--' b for blue, . is marker, -- is dotted line.
plt.plot(x2[:6], x2[:6]**2,'r', label='x**2', )
# plotting remaining part of graph
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.legend()

plt.savefig('matgraph.png', dpi=300)

plt.show()
```
